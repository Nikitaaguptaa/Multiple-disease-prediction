```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

%matplotlib inline
```

```python
parki_data = pd.read_csv("parkinsons.csv")
```

```python
parki_data.head()
```

```python
parki_data.info()
```

```python
parki_data['status'].value_counts()
```

```python
parki_data.groupby('status').mean(numeric_only=True)
```

```python
sns.pairplot(parki_data, hue="status")
```

```python
X = parki_data.drop('status', axis=1)
Y = parki_data['status']
```

```python
X.head()
```

```python
Y.head()
```

```python
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
Y_train.count()
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
parki_data = parki_data.drop(columns='name')
```

```python

classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
#Accuracy score on the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
#Accuracy score on the training data
X_train_prediction=classifier.predict(X_test)
training_data_accuracy = accuracy_score(X_train_prediction, Y_test)
```

```python
training_data_accuracy
```
